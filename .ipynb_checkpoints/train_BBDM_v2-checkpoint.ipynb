```python
from datasets.custom import CustomAlignedDataset

class DatasetConfig:
    def __init__(self, dataset_path, image_size, channels, to_normal, flip):
        self.dataset_path = dataset_path
        self.image_size = image_size
        self.channels = channels
        self.to_normal = to_normal
        self.flip = flip
        
dataset_config = DatasetConfig(
    dataset_path='brain_data',
    image_size=256,
    channels=1,
    to_normal=True,
    flip=False
)

# CustomAlignedDataset 클래스 로드
dataset = CustomAlignedDataset(dataset_config, stage='train')

# 데이터셋의 길이 확인
print(f"Dataset length: {len(dataset)}")

# 데이터셋에서 샘플 하나 확인
sample_ori, sample_cond = dataset[0]
print(f"Original image shape: {sample_ori[0].shape}")
print(f"Conditioned image shape: {sample_cond[0].shape}")
print(f"Original image type: {type(sample_ori[0])}")
print(f"Conditioned image type: {type(sample_cond[0])}")
```

```python
image_name = sample_ori[1]
print(image_name)
np_image = aaa
```

```python
if image_name.startswith('cbct'):
    # cbct norm -> [0,1]
    if image_name.startswith('cbct_2BA'):
        cbct_max, cbct_min = 3000.0, 0
    elif image_name.startswith('cbct_2BB'):
        cbct_max, cbct_min = 2000.0, -1000.0
    elif image_name.startswith('cbct_2BC'):
        cbct_max, cbct_min = 3000.0, -1024.0
    np_image =  (np_image - cbct_min) / (cbct_max - cbct_min)
elif image_name.startswith('ct'):
    # ct norm -> [0,1]
    np_image = (np_image - (-1024)) / (3000 - (-1024))
else: 
    raise ValueError("Unsupported image type.")
    
import PIL.Image as Image
if np_image.ndim == 2:
    np_image = np.stack([np_image]*3, axis=-1)
# image = Image.fromarray(np_image) 

# if not image.mode == 'RGB':
#     image = image.convert('RGB')
    
import torchvision.transforms as transforms   

p = 1.0
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.RandomHorizontalFlip(p=p),
    transforms.Resize((256,256))
   ])

image = transform(np_image) 

image = (image - 0.5) * 2.
```

```python
# print(np.max(image), np.min(image))
print(torch.max(image), torch.min(image))
```

```python
plt.imshow(image[0,:,:])
```

```python
from torchvision.utils import make_grid
@torch.no_grad()
def get_image_grid(batch, grid_size=4, to_normal=True):
    batch = batch.detach().clone()
    image_grid = make_grid(batch, nrow=grid_size)
    if to_normal:
        image_grid = image_grid.mul_(0.5).add_(0.5).clamp_(0, 1.)
    image_grid = image_grid.mul_(255).add_(0.5).clamp_(0, 255).permute(1, 2, 0).to('cpu', torch.uint8).numpy()
    return image_grid

image_tensor = [image, image, image, image ]
image_tensor = torch.stack(image_tensor)  # 원본 이미지들을 하나의 텐서로 결합
print(image_tensor.shape)  # 텐서의 형태 확인

image_grid = get_image_grid(image_tensor, grid_size=4)
```

```python
im = Image.fromarray(image_grid)

im.save('image_grid_sample.png')
im.show()  # 이미지를 화면에 표시
```
